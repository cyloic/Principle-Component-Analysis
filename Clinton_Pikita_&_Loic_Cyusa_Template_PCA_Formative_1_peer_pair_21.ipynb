### Step 1: Load and Standardize the Data
Before applying PCA, we must standardize the dataset. Standardization ensures that all features have a mean of 0 and a standard deviation of 1, which is essential for PCA.
Fill in the code to standardize the dataset.

```python
from google.colab import files
uploaded = files.upload()  # This will prompt you to upload your file

import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('fuel_econ.csv')

# Select only numeric columns for standardization
numeric_data = data.select_dtypes(include=[np.number]).to_numpy()  # Convert to NumPy array

# Handle missing values (replace NaNs with column mean)
nan_mask = np.isnan(numeric_data)  # Identify NaNs
column_means = np.nanmean(numeric_data, axis=0)  # Compute mean ignoring NaNs
numeric_data[nan_mask] = np.take(column_means, np.where(nan_mask)[1])  # Replace NaNs with column means

# Standardize the numeric data (Mean = 0, Std = 1)
mean = np.mean(numeric_data, axis=0)
std_dev = np.std(numeric_data, axis=0)
standardized_data = (numeric_data - mean) / std_dev

# Display the first few rows of standardized data
print(standardized_data[:5])
```

### Step 3: Calculate the Covariance Matrix
The covariance matrix helps us understand how the features are related to each other. It is a key component in PCA.

```python
# Step 3: Calculate the Covariance Matrix
import numpy as np

# Compute the covariance matrix using NumPy
cov_matrix = np.cov(standardized_data, rowvar=False)  # rowvar=False to treat columns as variables

# Display the covariance matrix
print("Covariance Matrix:\n", cov_matrix)
```

### Step 4: Perform Eigendecomposition
Eigendecomposition of the covariance matrix will give us the eigenvalues and eigenvectors, which are essential for PCA.
Fill in the code to compute the eigenvalues and eigenvectors of the covariance matrix.

```python
# Step 4: Perform Eigendecomposition
import numpy as np

# Assuming 'covariance_matrix' is already calculated from Step 3

# Perform eigendecomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Display the eigenvalues and eigenvectors
print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

### Step 5: Sort Principal Components
Sort the eigenvectors based on their corresponding eigenvalues in descending order. The higher the eigenvalue, the more important the eigenvector.
Complete the code to sort the eigenvectors and print the sorted components.

```python
# Step 5: Sort Principal Components

import numpy as np

# Sort the eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]  # Get indices for sorting (descending order)

# Sort eigenvalues and eigenvectors accordingly
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]  # Columns are eigenvectors

# Display the sorted eigenvalues and eigenvectors
print("Sorted Eigenvalues:\n", sorted_eigenvalues)
print("\nSorted Eigenvectors:\n", sorted_eigenvectors)
```

### Step 6: Project Data onto Principal Components
Now that weâ€™ve selected the number of components, we will project the original data onto the chosen principal components.
Fill in the code to perform the projection.

```python
# Step 6: Project Data onto Principal Components

import numpy as np

# Decide on the number of principal components to keep (choose manually or based on explained variance)
num_components = 2  # Example: Keep the top 2 principal components

# Select the top `num_components` eigenvectors (columns)
top_eigenvectors = sorted_eigenvectors[:, :num_components]

# Project the standardized data onto the selected principal components
reduced_data = np.dot(standardized_data, top_eigenvectors)

# Display the first few rows of projected data
print("Reduced Data (Projected onto Principal Components):\n", reduced_data[:5])
```

### Step 7: Output the Reduced Data
Finally, display the reduced data obtained by projecting the original dataset onto the selected principal components.

```python
# Step 7: Output the Reduced Data

# Step 7: Output the Reduced Data
print(f"Reduced Data Shape: {reduced_data.shape}")  # Display the shape of the reduced dataset

# Display the first 5 rows of the reduced data
print("\nFirst 5 Rows of Reduced Data:\n", reduced_data[:5])
```

### Step 8: Visualize Before and After PCA
Now, let's plot the original data and the data after PCA to compare the reduction in dimensions visually.

```python
# Step 8: Visualize Before and After PCA

import matplotlib.pyplot as plt

# Select first two numeric features from the original standardized data for comparison
original_x = standardized_data[:, 0]  # First feature
original_y = standardized_data[:, 1]  # Second feature

# Extract PCA-reduced components
pca_x = reduced_data[:, 0]  # First principal component
pca_y = reduced_data[:, 1]  # Second principal component

# Create a figure with two subplots (before and after PCA)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Original Data (First Two Features)
axes[0].scatter(original_x, original_y, alpha=0.5, color="blue")
axes[0].set_title("Original Data (First Two Features)")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Plot PCA Reduced Data (First Two Principal Components)
axes[1].scatter(pca_x, pca_y, alpha=0.5, color="red")
axes[1].set_title("PCA Reduced Data (First Two Components)")
axes[1].set_xlabel("Principal Component 1")
axes[1].set_ylabel("Principal Component 2")

# Show the plots
plt.tight_layout()
plt.show()
```
